```python
import matplotlib.pyplot as plt 
import numpy as np 
v1 = 10
v2 = 20
m = np.linspace(1,10, 100)
ke1 = (m*v1**2)/2
ke2 = (m*v2**2)/2
plt.figure(figsize=(10, 6))
plt.plot(m, ke1, 'b-', label=f'Velocity:{v1}m/s')
plt.plot(m, ke2, 'r-', label=f'Velocity:{v2}m/s')
plt.xlabel('Mass(kg)')
plt.ylabel('Kinetic Energy(Joules)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
